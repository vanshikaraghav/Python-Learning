```python
def stnd_fuc(mu, sigma, x):
  return 1/(sigma*np.sqrt(2*np.pi))*np.exp(-(x-mu)**2/(2*sigma**2))
```

```python
mu=100; sigma=10; sample=1000
pdf= np.random.normal(mu, sigma, sample)
bins=np.linspace(60, 140, 80)
fig, ax=plt.subplots()
ax.hist(pdf, bins,color='g')
ax.plot(bins, stnd_fuc(mu, sigma, bins), 'r')
plt.show()
```

```python
class FourCalculations:
    def __init__(self,first,second):
      self.first=first
      self.second=second
      if second==0:
        raise ValueError('the number should not be zero')
    def add(self):
      return self.first+self.second
    def sub(self):
      return self.first-self.second
    def mul(self):
      return self.first*self.second
    def div(self):
      return self.first/self.second
```

```python
a= FourCalculations(10,3)
print(f'First value is{a.first}\n'+f'Second value is{a.second}')
print(f'Aaddition: {a.add()}')
print(f'Subtraction: {a.sub()}')
print(f'Multiplication: {a.mul()}')
print(f'Division: {a.div():.2f}')
```

```python
FourCalculations.mul(a)
```

```python
b= FourCalculations(6,2)
print(f'First Value is {b.first}\n'+f'Second Value is {b.second}')
print(f'Addition: {b.add()}')
print(f'Subtraction: {b.sub()}')
print(f'Multiplication: {b.mul()}')
print(f'Division: {b.div():.2f}')
```

```python
c=FourCalculations(12,3)
print(f'First Value is {c.first}\n'+f'Second Value is {c.second}')
print(f'Addition: {c.add()}')
print(f'Subtraction: {c.sub()}')
print(f'Multiplication: {c.mul()}')
print(f'Division: {c.div():.2f}')
```

```python
oblist =[a,b,c]
for i in oblist:
  print(i.mul(), end=',')
```

```python
class FiveCalculation(FourCalculations):
  def __init__(self,first,second,third):
    super().__init__(first,second)
    self.third=third
  def pow(self):
    return self.first**self.second+self.third
  def div(self):
    if self.second==0:
      raise 0
    else:
      return (self.first/self.second)
```

```python
d=FiveCalculation(9,2,3)
print(d.pow(),d.add(),d.sub(), d.mul())
```

```python
#Another example
class Person:
  def __init__(self, first, last):
    self.firstname=first
    self.lastname=last
  def __repr__(self):
    return(self.firstname+' '+self.lastname)
```

```python
class Employee(Person):
  def __init__(self, first, last, staffnum):
    super().__init__(first, last)
    self.staffnumber=staffnum
  def __repr__(self):
    return (f'staff Name:{super().__repr__()},\n'+ f'Staff Num:{self.staffnumber}')
```

```python
x= Person("Youngsun","Kwon")
repr(x)
```
